# Create Model

```python
import blinkingModel as bl
model = bl.QsmlmModel()
```

# Set Parameters
Set values for the blinking model parameters (p and q) and a valid state vector(s), as well as a valid initial weight matrix (w). s and w must have identical dimensions. 

```python
p = 0.282
q = 0.29
s = [0,1,2,5,8]
w = [1,1,1,1,1]

model.setModelParameters(p,q,s,w)
```

# Load qSMLM Blinking Statistic Data Set
Define the columns to be loaded

```python
x=5
y=6
model.loadFile(x,y)
```

# Data Modeling

Model parameter estimation using a Trust Region Reflective algorithm  based least squares routine

```python
model.lsOptimization()
```

Model parameter estimation using a L-BFGS-B based maximum Likelihood estimation routine

```python
model.mleOptimization()
```

# Plot Modeling Results

Plot the LS-based model

```python
model.lsPlot()
```

Plot the MLE-based model

```python
model.mlePlot()
```

# Save Model

Save the LS-based model

```python
model.lsSaveResults()
```

Save the MLE-based model

```python
model.mleSaveResults()
```
